```python
df = df.set_index('ID') 
```

```python
#Realizar un proceso de limpieza del dataset
#Drop unamed column
df = df.drop('Unnamed: 0',axis=1) 
df.head()
```

```python
#Find if exists columns with  NaN values, list them and replace values with 0
df.columns[df.isna().any()].tolist()
```

```python
#create list with non object columns 
positions = ['LS',
 'ST',
 'RS',
 'LW',
 'LF',
 'CF',
 'RF',
 'RW',
 'LAM',
 'CAM',
 'RAM',
 'LM',
 'LCM',
 'CM',
 'RCM',
 'RM',
 'LWB',
 'LDM',
 'CDM',
 'RDM',
 'RWB',
 'LB',
 'LCB',
 'CB',
 'RCB',
 'RB',
 ]

#clean NaN in dataframe
def fill_na(df):
  for col in df:
    if (df.dtypes[col] == np.int64):
          df[col].fillna(0,inplace=True)
    elif (df.dtypes[col] ==np.float64):
          df[col].fillna(0,inplace=True)
    else: 
      if df[col].name in positions:
        df[col].fillna(0,inplace=True)
        df[col].replace('nan',0,inplace=True)
      else:
        df[col].replace(np.nan,"Undefined",inplace=True)

fill_na(df)
```

```python
#Sum values in position columns
def sum_pos(df):
  for col in df: 
    if df[col].name in positions:
      df[col + "_unit"]=df[col].map(pd.eval)
sum_pos(df)
```

```python
#Drop columns
def drop_columns (df):
  for col in df: 
    if df[col].name in positions:
      df.drop(col,axis=1)
drop_columns(df)

df.head(11)
```

```python
#extraer valores de las columnas tipo object para convertirlas en numeros
df['Value'] = df['Value'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int) #Extract value 
df['Wage'] = df['Wage'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int) #Extract wage
df['Release Clause'] = df['Release Clause'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int) #Extract release clause value
```

```python
fill_na(df)
```

```python
df['Value_M']=df['Value']*100000
df['Wage_K']=df['Wage']*1000
df['Release_M']=df['Release Clause']*100000
```

```python
df['Value_M']=df['Value_M'].astype(float)
df['Wage_K']=df['Wage_K'].astype(float)
df['Release_M']=df['Release_M'].astype(float)
df.head
```
